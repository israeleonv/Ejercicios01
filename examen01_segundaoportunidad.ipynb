```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Ejercicio de gr√°ficas
#inciso a
y= lambda x: 1 - np.exp(x) + (np.exp(1) - 1)*np.sin(np.pi*x/2)
x = np.linspace(-1,0,100)
plt.plot(x,y(x))
plt.grid()
```

```python
#inciso b
y= lambda x: (x-1)*np.tan(x)+ x * np.sin(np.pi * x)
x = np.linspace(-1,0,100)
plt.plot(x,y(x))
plt.grid()
```

```python
#inciso c
y= lambda x: x * np.sin(np.pi * x)-(x-2)* np.log(x)
x = np.linspace(0.001,1,100)
plt.plot(x,y(x))
plt.grid()
```

```python
#inciso c
y= lambda x: (x-2)*np.sin(x)*np.log(x+2)
x = np.linspace(-1,0,100)
plt.plot(x,y(x))
plt.grid()
```

```python
#Ejercicio del polinomio de Taylor
#Aproximacion A
def aproxA (x,n):
    suma=0
    for i in range(0,n+1):
        suma += (-1)**i * 5**i/ np.math.factorial(i)
    return suma
x = -5
n = 21
e1 = aproxA(x,n)

print('e = {:.55f}'.format(np.exp(-5)))
print('e = {:.55f}'.format(e1))
```

```python
#Aproximacion A
def aproxB (x,n):
    suma=0
    for i in range(0,n+1):
        suma += (5)**i / np.math.factorial(i)
    return 1/suma
x = -5
n = 14
e2 = aproxB(x,n)

print('e = {:.55f}'.format(np.exp(-5)))
print('e = {:.55f}'.format(e2))
```
